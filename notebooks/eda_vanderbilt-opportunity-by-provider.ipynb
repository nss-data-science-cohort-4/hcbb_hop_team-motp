```python
npi_query = """
    SELECT n.npi, n.provider_organization_name, s.Classification AS specialty

    FROM nppes AS n

    JOIN cbsa AS c
        ON c.ZIP = CAST(n.zip_5 AS INTEGER)
    
    LEFT JOIN specialty AS s
        ON s.npi = n.npi
        
    WHERE n.entity_type_code = 2
    --AND LOWER(s.Classification) LIKE '%hospital%'
    AND LOWER(n.provider_organization_name) LIKE '%centennial%'
"""
org_data = pd.read_sql(npi_query, db)
```
