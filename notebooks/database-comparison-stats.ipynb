```python
nppes_specialty = """
SELECT n.*, s.Classification
FROM nppes AS n
LEFT JOIN specialty AS s
    ON s.npi = n.npi
"""
```

```python
nppes_specialty = pd.read_sql(nppes_specialty, db)
```

```python
nppes_specialty[nppes_specialty['Classification'].isna() == True].info()
```

```python
nppes_zip_specialty = """
SELECT n.*, s.Classification
FROM nppes AS n
JOIN cbsa AS c
    ON c.ZIP = CAST(n.zip_5 AS INTEGER)
LEFT JOIN specialty AS s
    ON s.npi = n.npi
"""
```

```python
nppes_zip_specialty = pd.read_sql(nppes_zip_specialty, db)
```

```python
nppes_zip_specialty.info()
```
